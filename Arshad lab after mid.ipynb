```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cubicinterpolation():
    x_pts = np.linspace(0,2*np.pi,10) #10 number of points(equidistant) #domain points
    y_pts = np.sin(x_pts)
    x_vals = np.linspace(0,2*np.pi,50) #jo function interpolate krna hai uski domain hai
    
    f = interpolate.interp1d(x_pts,y_pts,'cubic')
    y_vals = f(x_vals) #cubic interpolation, main program
    
    plt.plot(x_pts,y_pts,'o')#plot known data points
    plt.plot(x_vals,y_vals,'-x') #plot interpolated points
    plt.show()
```

```python
cubicinterpolation()
```
